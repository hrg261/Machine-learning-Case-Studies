## PART 3 : FIND 1000 CUSTOMERS WHICH ARE INVOLVED IN MALICIOUS ACTIITIES 

```python
data_history=pd.read_csv("7. HEALTH CARE CASE STUDY - CLOUDERA CHALLENGE_2//Files//Patient_history_samp.csv")
data_transct=pd.read_csv("7. HEALTH CARE CASE STUDY - CLOUDERA CHALLENGE_2//Files//Transaction_coo.csv")
```

```python
data_history.head()
```

```python
data_history.info()
```

```python
data_history.age=data_history.age.fillna(data_history.age.mode()[0])
data_history.income=data_history.income.fillna(data_history.income.mode()[0])
```

```python
data_history.info()
```

```python
data_transct.head()
```

```python
data_transct.info()
```

```python
data_transct[data_transct.duplicated('id')==True].head()
```

```python
data_transct=data_transct[['id','count']].groupby('id').sum()
```

```python
data_transct.head()
```

```python
data_history.set_index('id',inplace=True)
```

```python
data_history.head()
```

```python
data_patients=pd.concat([data_transct,data_history],join='inner',axis=1)
```

```python
data_patients.head()
```

```python
data_patients.info()
```

```python
data_patients.head()
```

```python
# An utility function to create dummy variable
def create_dummies( df, colname ):
    col_dummies = pd.get_dummies(df[colname], prefix=colname)
    col_dummies.drop(col_dummies.columns[0], axis=1, inplace=True)
    df = pd.concat([df, col_dummies], axis=1)
    df.drop( colname, axis = 1, inplace = True )
    return df
```

```python
dummy=['age','gender','income']
for var in dummy:
    data_patients[var] = data_patients[var].astype('category')
    data_patients=create_dummies(data_patients,var)
```

```python
data_patients.head()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
data_patients_scaled=scaler.fit_transform(data_patients)
```

```python
data_patients2=pd.DataFrame(data_patients_scaled)
data_patients2.columns=data_patients.columns
data_patients2.index=data_patients.index
```

```python
data_patients2.head()
```

```python
from sklearn.neighbors import LocalOutlierFactor
```

```python
anamoly=LocalOutlierFactor(n_neighbors=40,n_jobs=-1,contamination=.002)
```

```python
a=anamoly.fit_predict(data_patients2)
```

```python
label=pd.DataFrame(a)
label.index=data_patients2.index
label.columns=['label']
```

```python
label.label.value_counts()
```

```python
patients_outlier=data_patients2[label.label==-1]
```
